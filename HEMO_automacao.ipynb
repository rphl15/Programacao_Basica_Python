*Remover linhas que não contem resposta*

```python
condicao = (
    df['Tipo Anestesia'].str.strip().replace('', np.nan).isna() &
    df['Setor/Sala'].str.contains('Hemodinâmica', case=False, na=False)
)
df.loc[condicao, 'Tipo Anestesia'] = 'Sedação'
df[condicao]
```

*Filtrando os dados*

```python
palavras = (
    'Local',
    'Local Assistida',
    'Sem analgesia'   
)

padrao = '|'.join(palavras)

vazios = (
    (df['Tipo Anestesia'].fillna('').str.strip().eq('')) |
    (df['Tipo Anestesia'].str.contains(padrao, case=False, na=False)) &
    df['Setor/Sala'].str.contains('Hemodinâmica', case=False, na=False)
)
df.loc[vazios, 'Cor'] = 'Amarelo'
df[vazios]
```
