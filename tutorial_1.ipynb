*Usando if, else e and*

```python
nota = 7
frequencia = 90
if nota >= 7 and frequencia > 70:
    print("Aprovado")
else:
    print("Reprovado")
```

*Usando if, else e or*

```python
nota = 6
frequencia = 60
if nota >= 7 or frequencia > 70:
    print("Aprovado")
else:
    print("Reprovado")
```

*Utilizando elif*

```python
nota = 5
if nota <= 4:
    print("Reprovado")
elif nota > 4 and nota <=6:
    print("Exame")
else:
    print("Aprovado")
```
