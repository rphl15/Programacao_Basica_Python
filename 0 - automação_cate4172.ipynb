*Importando o caminho*

```python
mes = 'MAR'
local_exportação = 'G:/Drives compartilhados/Qualidade/ACURÁCIA/CATE 4712 ATB/2026/'+mes+'/'
caminho = 'G:/Drives compartilhados/Qualidade/ACURÁCIA/CATE 4712 ATB/2026/'+mes+'/'
hospital = 'Mat_STAR'
```

*Importando bibliotecas*

```python
import pandas as pd
import numpy as np
import openpyxl
import time

# Repositorio - Diretorio
import shutil
import os
```

*Trazendo o arquivo - NÃO ESQUEÇA DE ALTERAR A UNIDADE*

```python
df = pd.read_excel('G:/Drives compartilhados/Qualidade/ACURÁCIA/CATE 4712 ATB/2026/' + mes + '/CATE_4712_' + hospital + '_' + mes + '.xlsx')
```

*Filtrando dados*

```python
#Alterando a coluna com tipo Time
df['Inicio Anestesia'] = pd.to_datetime(df['Inicio Anestesia'], errors='coerce')
df['Inicio do Procedimento cirurgico '] = pd.to_datetime(df['Inicio do Procedimento cirurgico '], errors='coerce')

rd = df.loc[df['Inicio Anestesia'] <= df['Inicio do Procedimento cirurgico ']]

rd[['Inicio Anestesia','Inicio do Procedimento cirurgico ']]

kd = df.loc[df['Inicio Anestesia'] > df['Inicio do Procedimento cirurgico ']]

#Retirando espaços da linha
df['Inicio Anestesia'] = df['Inicio Anestesia'].replace(r'^\s*$', np.nan, regex=True)

#Filtrando linhas que está sem dados
nd = df[df['Inicio do Procedimento cirurgico '].isnull()]
ld = df[df['Inicio Anestesia'].isnull()]
```

*Criando colunas e preenche*

```python
rd['Tempo decorrido'] = 'Procedimento iniciou após Anestesia'
kd['Tempo decorrido'] = 'Procedimento iniciou antes ou durante a Anestesia'
nd['Tempo decorrido'] = 'Revisar atendimento devido a falta de registros'
ld['Tempo decorrido'] = 'Revisar atendimento devido a falta de registros'
```

*Concatenando dataframes*

```python
frames = [kd, rd, ld, nd]
df = pd.concat(frames)
```

*Filtrando ATBs*

```python
df['Apl_Cefazolina'] = 'Conforme'
df['Apl_tempo_240m'] = 'Conforme'

cf = df.loc[df['Medicamento'].str.contains('CEFAZOLINA')]

df['Peso'] = pd.to_numeric(df['Peso'], errors ='coerce')
cf = df.loc[df['Peso'] >= 120 ]

cf['Dose'] = pd.to_numeric(cf['Dose'], errors = 'coerce')
cf = cf.loc[cf['Dose'] < 3 ]

cf['Apl_Cefazolina'] = 'Não Conforme (PC 120Kg e com menos de 3g de ATB aplicada)'

#Atulizando datraframe
df.update(cf)
```

*Filtrando Repiques*

```python
df['Tempo Procedimento'] = pd.to_numeric(df['Tempo Procedimento'], errors = 'coerce')
md = df.loc[df['Tempo Procedimento'] > 240]

df['Tempo Inicio Atb x Inicio Procedimento'] = pd.to_numeric(df['Tempo Inicio Atb x Inicio Procedimento'], errors = 'coerce')
md = md.loc[md['Tempo Inicio Atb x Inicio Procedimento'] <= -180]

df['Apl_tempo_240m'] = df['Apl_tempo_240m'].astype('object')
md['Apl_tempo_240m'] = md['Apl_tempo_240m'].astype('object')

md['Apl_tempo_240m'] = 'Não conforme'

#Atulizando datraframe
df.update(md)
```

*Exportando planilha pro excel*

```python
df.to_excel(caminho + 'CATE_4712_' + hospital + '_' + mes + '(FINALIZADO).xlsx')
print('-' * 20)
print("FINALIZADO")
print('-' * 20)
```
