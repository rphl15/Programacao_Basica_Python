```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import os
import regex
import unicodedata
```

```python
# Mostrando os caminhos dos arquivos
caminho = "G:/Drives compartilhados/Qualidade/ACURÁCIA/CATE 3136/2026/03 - MAR/"
arquivo = "CATE3136_ITAIM_MAR26.xlsx"

caminho_completo = os.path.join(caminho, arquivo)

df = pd.read_excel(caminho_completo)
print(df.shape)
```

```python
# Exibindo as colunas
df.columns
```

```python
# Filtrando primeiro data frame
hemo = df['Setor/Sala'].str.contains('Hemodinâmica' or 'Centro Obstétrico', na=True)
df_hemo = df[hemo]
df_con = df[~hemo]
print(df_hemo.shape)
print(df_con.shape)
print(df.shape)
```

```python
# Colocando a cor verde
def cor_linha_green(row):
    if row['Setor/Sala'] in ['Hemodinâmica','Centro Obstétrico']:
        return ['background-color: green'] * len(row)
    return [''] * len(row)

df_con.style.apply(cor_linha_green, axis=1)
```

```python
# Filtrando linhas
print(df.shape)

filtro_igual = df_con['Tipo Anestesia'].isin(['Local','Local Assistida','Sem analgesia'])
filtro_final = filtro_igual

df_removido = df_con.loc[filtro_final]
df_con = df_con.loc[~filtro_final]

print(df_removido.shape)
print(df_con.shape)
```

```python
# Ver as cores
def cor_linha(row):
    if row['Tipo Anestesia'] in ['Local','Local Assistida','Sem analgesia']:
        return ['background-color: yellow'] * len(row)
    return [''] * len(row)

df_con.style.apply(cor_linha, axis=1)
```

```python
# Filtrando linhas
vazia = df_con[df_con["Tipo Anestesia"].isna() | (df_con["Tipo Anestesia"].str.strip() == "")]
print(vazia.shape)
```

```python
# Selecionando as palavras
import unicodedata

def remover_acentos(texto):
    if isinstance(texto, str):
        return ''.join(
            c for c in unicodedata.normalize('NFKD', texto)
            if not unicodedata.combining(c)
        )
    return texto
palavras = [
    "postectomia",
    "vasectomia",
    "retirada de cateter",
    "retirada de duplo j",
    "extensos ferimentos",
    "endoscopia",
    "colonoscopia",
    "infiltracao",
    "denervacao",
    "bloqueio",
    "tunel",
    "exerese",
    "aspiracao manual"   
]
certo = vazia['Ds procedimento'] \
    .apply(remover_acentos) \
    .str.lower() \
    .str.contains("|".join(palavras), na=False)
filtrado = vazia[certo] 

vazia.loc[certo, 'Tipo Anestesia'] = 'Anestesia'
anestesia = vazia[certo]
print(anestesia.shape)
print(vazia.shape)
```

```python
# Concatenando dataframes
df_final = (
    pd.concat([vazia, anestesia, df_con, df_removido, df_hemo])
    .drop_duplicates(subset=['Ds procedimento', 'þÿNr atendimento'], keep='last')
)
```
