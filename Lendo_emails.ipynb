*Importando as bibliotecas*

```python
import pyautogui as pya
import time
```

*Movendo o mouse para abrir uma nova janela*

```python
pya.position()
pya.moveTo(367, 67)
pya.click(367, 67)
pya.hotkey('ctrl', 'a')
pya.press('delete')
```

*Escrever o "gmail"*

```python
pya.typewrite("https://mail.google.com/mail/u/0/#inbox")
pya.press('enter')
```

*Abrir um email*

```python
pya.position()
time.sleep(3)
pya.click(528, 385)
```

*Ir para proximo email*

```python
pya.click(919, 214, clicks=10000, interval=0.1)
```
