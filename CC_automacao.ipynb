*Remover linhas que não contem resposta*

```python
condicao = (
    df['Setor/Sala'].isnull() &
    df['Tipo Anestesia'].str.contains(
        'Local|Local Assistida|Sem Analgesia',
        case=False,
        na=False
    )
)
df.loc[condicao, 'Cor'] = 'Amarelo'
print(df.loc[condicao, 'Cor'])
```

*Filtrando os dados*

```python
palavras = (
    'Postectomia',
    'Vasectomia',
    'Retirada de Cateter',
    'Retirada de duplo j',
    'Entensos Ferimentos',
    'Endoscopia',
    'Colonoscopia',
    'Infiltração',
    'Denervação',
    'Bloqueio',
    'Túnel',
    'Exérese',
    'Aspiração Manual'    
)

padrao = '|'.join(palavras)

vazios = (
    df['Tipo Anestesia'].isnull() &
    df['Ds procedimento'].str.contains(padrao, case=False, na=False) &
    df['Proced adicional'].isnull()
)
df.loc[vazios, 'Tipo Anestesia'] = 'Sedação'
```

```python
# Adicionando "Anestesia" em campos vazios
df.loc[
    df['Tipo Anestesia'].isnull() |
    (df['Tipo Anestesia'].str.strip() == ''),
    'Tipo Anestesia'
] = 'Anestesia'
```
