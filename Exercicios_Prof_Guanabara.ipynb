*Exercício 006 - sem importar bibliotecas*

```python
d1 = int(input('Digite um número: '))
db = d1 * 2
tr = d1 * 3
rz = d1 ** 0.5
print(f'Seu número é {d1} e seu dobro é {db} o triplo é {tr} e a raiz quadrada é {rz}')
```

*Exercício 006 - Importando bibliotecas*

```python
import math

d1 = int(input('Digite um número: '))
db = d1 * 2
tr = d1 * 3 
rz = math.sqrt(d1)
print(f'Seu número é {d1} e seu dobro é {db} o triplo é {tr} e a raiz quadrada é {rz}')
```
